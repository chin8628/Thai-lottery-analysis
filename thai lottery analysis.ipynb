```python
import os
import pandas as pd
import numpy as np

LOTTO_DIR = './lottonumbers/'
```

```python
lotto_files = os.listdir(LOTTO_DIR)

lotto_files.sort()
print(lotto_files[0], lotto_files[len(lotto_files) - 1])

lotto_files = pd.Series(lotto_files)
```

```python
dataframe_lottos = {}

for lotto_file in lotto_files:
    with open(LOTTO_DIR + lotto_file, 'r') as fp:
        data = fp.read()

    dataframe = []
    index = []

    #skip url line by slicing index
    for line in data.split('\n')[1:]:
        splited = line.split(' ')
        index.append(splited[0])

        dataframe.append(splited[1:])
        
    dataframe_lottos[lotto_file] = pd.DataFrame(dataframe, index=index).T
```

```python
# วิเคราะห์ เลขสองตัวท้าย 

last_two_list = []
for key in dataframe_lottos:
    last_two = int(dataframe_lottos[key]['TWO'][0])
    last_two_list.append(last_two)
    
last_two_list = pd.Series(last_two_list)

print(last_two_list.value_counts().head(10))
```

```python
two_reward = [85, 5, 79]
last_two_reward_date = []

for key in dataframe_lottos:
    last_two = int(dataframe_lottos[key]['TWO'][0])
    if last_two in two_reward:
        last_two_reward_date.append((last_two, key))
        
sorted(last_two_reward_date)
```

```python
# วิเคราะห์ รางวัลสามตัวท้าย

last_three_list = []
for key in dataframe_lottos:
    calling_index = 'THREE_LAST' if 'THREE_LAST' in dataframe_lottos[key].keys() else 'THREE'
    number_reward = 2 if 'THREE_LAST' in dataframe_lottos[key].keys() else 4
    
    for idx in range(number_reward):
        three = int(dataframe_lottos[key][calling_index][idx])
        last_three_list.append(three)
        
last_three_list = pd.Series(last_three_list)

print(last_three_list.value_counts().head(10))
```

```python
# วิเคราะห์ รางวัลที่ห้า

fifth_list = []
for key in dataframe_lottos:
    for idx in range(100):
        fifth = int(dataframe_lottos[key]['FIFTH'][idx])
        fifth_list.append(fifth)
        
fifth_list = pd.Series(fifth_list)

print(fifth_list.value_counts().head(10))
```

```python
# วิเคราะห์ รางวัลที่หนึ่ง

first_list = []
for key in dataframe_lottos:
    if dataframe_lottos[key]['FIRST'][0] != '':
        first = int(dataframe_lottos[key]['FIRST'][0])
        first_list.append(first)
        
first_list = pd.Series(first_list)

print(first_list.value_counts().head(10))
```
