```python
import os
import pandas as pd
import numpy as np

LOTTO_DIR = './lottonumbers/'
```

```python
lotto_files = os.listdir(LOTTO_DIR)
lotto_files = pd.Series(lotto_files)

lotto_files.head()
```

```python
dataframe_lottos = {}

for lotto_file in lotto_files:
    with open(LOTTO_DIR + lotto_file, 'r') as fp:
        data = fp.read()

    dataframe = []
    index = []

    #skip url line by slicing index
    for line in data.split('\n')[1:]:
        splited = line.split(' ')
        index.append(splited[0])

        dataframe.append(splited[1:])
        
    dataframe_lottos[lotto_file] = pd.DataFrame(dataframe, index=index).T
```

```python
dataframe_lottos['2015-09-01.txt']
```

```python
# วิเคราะห์ เลขสองตัวท้าย 

last_two_list = []
for key in dataframe_lottos:
    last_two = int(dataframe_lottos[key]['TWO'][0])
    last_two_list.append(last_two)
    
last_two_list = pd.Series(last_two_list)

print("จำนวนงวด:", len(last_two_list))
print(last_two_list.value_counts().head(10))
```

```python
# วิเคราะห์ เลขสามตัวท้าย 

fifth_list = []
for key in dataframe_lottos:
    for idx in range(100):
        fifth = int(dataframe_lottos[key]['FIFTH'][idx])
        fifth_list.append(fifth)
        
fifth_list = pd.Series(fifth_list)

print(fifth_list.value_counts().head(10))
```
